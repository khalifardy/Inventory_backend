```python
class Geometri:
    
    #konstruktor
    def __init__(self,nama,desc):
        self.__nama = nama
        self.__desc = desc
    
    def getnama(self):
        return self.__nama
    
    def deskripsi(self):
        return self.__desc



    
```

```python
from abc import abstractmethod

class BangunDatar:
    @abstractmethod
    def luas(self):
        pass
    @abstractmethod
    def keliling(self):
        pass

class BangunRuang:
    @abstractmethod
    def luaspermukaan(self):
        pass
        
    def volume(self):
        pass
```

```python
import math
class Lingkaran(Geometri,BangunDatar):
    #kosntruktor
    def __init__(self,jari_jari,nama,desc):
        super().__init__(nama,desc)
        self.__jari_jari = jari_jari
        self.__pi = math.pi #final atribute
    
    def luas(self):
        return self.__pi * (self.__jari_jari)**2
    
    def keliling(self):
        return 2 * self.__pi * self.__jari_jari
    
    def getjarijari(self):
        return self.__jari_jari
```

```python
class Persegi(Geometri,BangunDatar):
    #konstruktor
    def __init__(self,nama,desc,sisi):
        super().__init__(nama,desc)
        self.__sisi = sisi
    
    def luas(self):
        return (self.__sisi)**2
    
    def keliling(self):
        return 4*self.__sisi
    
    def getsisi(self):
        return self.__sisi
```

```python
class PersegiPanjang(Geometri,BangunDatar):
    #konstruktor
    def __init__(self,nama,desc,panjang,lebar):
        super().__init__(nama,desc)
        self.__panjang = panjang
        self.__lebar = lebar
    
    def luas(self):
        return self.__panjang * self.__lebar
    
    def keliling(self):
        return 2 * (self.__panjang + self.__lebar)
    
    def getpanjang(self):
        return self.__panjang
    
    def getlebar(self):
        return self.__lebar
    
```

```python
class Segitiga(Geometri, BangunDatar):
    def __init__(self,nama,desc,alas,tinggi,sisi2,sisi3):
        super().__init__(nama,desc)
        self.__alas = alas
        self.__tinggi = tinggi
        self.__sisi2 = sisi2
        self.__sisi3 = sisi3
    
    def luas(self):
        return self.__alas * self.__tinggi *0.5
    
    def keliling(self):
        return self.__alas + self.__sisi2 + self.__sisi3
    
    def getalas(self):
        return self.__alas
    
    def gettinggi(self):
        return self.__tinggi
    
    def getsisi2(self):
        return self.__sisi2
    
    def getsisi3(self):
        return self.__sisi3
    
```

```python
class Tabung(Lingkaran,BangunRuang):
    #konstruktor
    def __init__(self,nama,desc,jari_jari,tinggi):
        super().__init__(jari_jari,nama,desc)
        self.__tinggi = tinggi
    
    def volume(self):
        return super().luas() * self.__tinggi
    
    def luaspermukaan(self):
        return 2*super().luas()+ super().keliling() * self.__tinggi
    

    
```

```python
class Kubus(Persegi,BangunRuang):
    #konstruktor
    def __init__(self,nama,desc,sisi):
        super().__init__(nama,desc,sisi)
        
    
    def volume(self):
        return super().getsisi()**3
    
    def luaspermukaan(self):
        return 6 * super().luas()
```

```python
class Balok(PersegiPanjang,BangunRuang):
    #konstruktor
    def __init__(self,nama,desc,panjang,lebar,tinggi):
        super().__init__(nama,desc,panjang,lebar)
        self.__tinggi = tinggi
    
    def volume(self):
        return super().luas() *self.__tinggi
    
    def luaspermukaan(self):
        return 2 * (super().luas()+super().getpanjang()*self.__tinggi + super().getlebar()*self.__tinggi)
```

```python
class Prisma(Segitiga,BangunRuang):
    def __init__(self,nama,desc,alas,tinggi,sisi2,sisi3,tinggi_prisma):
        super().__init__(nama,desc,alas,tinggi,sisi2,sisi3)
        self.__tinggi_prisma = tinggi_prisma
    
    def volume(self):
        return super().luas() * self.__tinggi_prisma
    
    def luaspermukaan(self):
        return 2 * super().luas() + self.__tinggi_prisma *(super().getalas() + super().getsisi2() + super().getsisi3())
```
